```python
%matplotlib inline
# import dependences
import pandas as pd
import pandas as pd
import scipy.stats as stats
import numpy as np
import matplotlib.pyplot as plt
import datetime 
```

```python
crime_data = "crime_data.csv"
crime_df = pd.read_csv(crime_data)

```

```python
drug_crime = crime_df.loc[crime_df["Category"] == "DRUG/NARCOTIC", :]

sample_years = ['2013','2014','2015', '2016', '2017']
five_years = crime_df[crime_df['Date'].str.contains('|'.join(sample_years))]


crime_drug_tenderloin = five_years[five_years["PdDistrict"] == "TENDERLOIN"]
crime_drug_tenderloin = crime_drug_tenderloin.set_index("Category")


search_for = ['COCAINE', 'MARIJUANA', 'HEROIN' 'METH']
drugs = crime_drug_tenderloin [crime_drug_tenderloin ["Descript"].str.contains('|'.join(search_for))]
drugs["year"] = pd.DatetimeIndex(drugs["Date"]).year
drugs.sort_values(by = "year", ascending = False)
```

```python
drugs["Resolution"].value_counts()
```

```python
#Drug incidents by year
drugs_year = drugs.groupby(["year"]).count()
drugs_year
```

```python
#Drug incidents by district
# arrests = drugs[drugs["Resolution"].str.contains("ARREST, BOOKED")]
arrests = drugs.loc[drugs["Resolution"] == "ARREST, BOOKED", :]
arrests_year = arrests.groupby(["year"]).count()
arrests_year.head()

```

```python
resolution = arrests_year["IncidntNum"]
```

```python
arrests_pct = round((arrests_year / drugs_year * 100),1)
arrests_pct_reset = arrests_pct.reset_index()
arrests_pct_reset
```

```python
sample_years = ['2013','2014','2015', '2016', '2017']
incidents = arrests_pct_reset["IncidntNum"]
(sample_years)
```

```python
x_axis = sample_years
y_axis = incidents
plt.xlabel("Year")
plt.ylabel("Arrest Rate (%)")
plt.title("Arrest Rate by Year")
plt.plot(x_axis, y_axis, marker = 'o', color="blue")

plt.savefig("Arrest Rate JK")

#sine_handle, = plt.plot(x_axis, sin, marker ='o', color='blue', label="Sine")
```
